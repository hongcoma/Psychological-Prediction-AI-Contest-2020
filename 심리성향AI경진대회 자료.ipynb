### 비 식별화를 위해 Secret 처리된 문항
- Never tell anyone the real reason you did something unless it is useful to do so.
- 당신이 어떤 일을 한 이유를 필요할 때를 제외하고는 누구에게도 절대 말하지 않는다.

- Honesty is the best policy in all cases.
- 모든 케이스에 있어서 정직이 최선이다.

- Generally speaking, people won't work hard unless they're forced to do so.
- 일반적으로, 사람들을 강제하지 않는 한, 열심히 일하지 않는다. 

- When you ask someone to do something for you, it is best to give the real reasons for wanting it rather than giving reasons which carry more weight.
- 어떤 일을 해달라고 누군가에게 요청할 때, 내가 그 일을 원하는 실제 이유를 알려주는 것이, 더 중요해보이는 듯한 이유를 알려주는 것보다 가장 좋은 방법이다.

- Most people who get ahead in the world lead clean, moral lives.
- 세상을 앞에서 이끌어가는 사람들은 깨끗하고 도덕적인 삶을 산다.

- Most people are brave.
- 대부분의 사람들은 용감하다.

- It is possible to be good in all respects.
- 모든 면에서 좋은 것이 가능하다. 

- People suffering from incurable diseases should have the choice of being put painlessly to death.
- 불치병에 걸린 사람들에게 안락사를 선택할 권리를 갖게 해야 한다.

# Feature 설명
1. Q_A / Q_E : 알파벳에 대한 응답과 응답까지 걸린 시간 (Q_A : 1 ~ 5 사이로 응답)
2. age_group : 연령
3. education : 교육 수준 
    - 1 : Less than high school : 중학교
    - 2 : High school : 고등학교
    - 3 : University degree : 대학교 졸업
    - 4 : Graduate degree : 대학원 졸업
4. engnat : 모국어가 영어
    - 1 : Yes
    - 2 : No
    - 3 : 무응답
5. familysize : 형제자매 수
6. gender : 성별
    - Male / Female
7. hand : 필기하는 손
    - 0 : 무응답
    - 1 : Right
    - 2 : Left
    - 3 : Both
8. married : 혼인상태
    - 0 : Other
    - 1 : Never married
    - 2 : Curently married
    - 3 : Previously married
9. race : 인종
    - Asian / Arab / Black / Indigenous australian / Native American / White / Ohter
10. religion : 종교
    - Agnostic / Atheist / Buddhist / Christian_Catholic / Christian_Mormon / Christian_Protestant / Christian_Other / Hindu / Jewish / Muslim / Sikh / Other
11. tp__ : 나는 다음 단어와 관련이 깊다 (1 ~ 7 사이로 응답)
12. urban : 유년기의 거주 구역
    - 0 : 무응답
    - 1 : Rural(country side
    - 2 : Suburban
    - 3 : Urban (town, city)
13. wr_(1~13) : 설문자의 어휘 능력 (실존하는 해당 단어의 정의를 앎)
14. wf_(1~3) : 설문자의 어휘 능력 (허구인 단어의 정의를 앎)
